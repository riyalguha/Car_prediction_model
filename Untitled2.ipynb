```python
import tensorflow as tf
import pandas as pd
import seaborn as sns
from tensorflow.keras.layers import Normalization,Dense,InputLayer
from tensorflow.keras.losses import MeanSquaredError,Huber,MeanAbsoluteError
from tensorflow.keras.optimizers import Adam
import matplotlib.pyplot as plt
from tensorflow.keras.metrics import RootMeanSquaredError
import numpy as np
```

### The features of the ML model are years,km,rating,condition,economy,top speed,hp,torque

## **Data** **Preparation**

```python
data = pd.read_csv("train.csv",",")
data.head()
```

```python
data.shape
```

```python
sns.pairplot(data[['years','km','rating','condition','economy','top speed','hp','torque']],diag_kind='kde')
```

```python
tensor_data = tf.constant(data)
print(tensor_data.shape)
```

```python
tensor_data = tf.cast(tensor_data,tf.float32)
print(tensor_data)
```

```python
tensor_data = tf.random.shuffle(tensor_data)
```

```python
print(tensor_data[:5])
```

```python
X=tensor_data[:,3:-1]
```

```python
y=tensor_data[:,-1]
```

```python
X[:5]
# print(X.shape)
```

```python
y[:5]
```

```python
y=tf.expand_dims(y,axis=-1)
print(y[:5])
```

```python
print(X.shape)
```

```python
TRAIN_RATIO=0.8
VALIDATION_RATIO=0.1
TEST_RATIO=0.1
DATASET_SIZE=len(X)
```

```python
X_train=X[:int(DATASET_SIZE*TRAIN_RATIO)]
y_train=y[:int(DATASET_SIZE*TRAIN_RATIO)]
print(X_train.shape)
print(y_train.shape)
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((X_train,y_train))
#if data if not already shuffled we can shuffle using this data API
train_dataset = train_dataset.shuffle(buffer_size=8,reshuffle_each_iteration=True).batch(32).prefetch(tf.data.AUTOTUNE)
```

```python
for x,y in train_dataset:
  print(x,y)
  break
```

```python
X_val=X[int(DATASET_SIZE*TRAIN_RATIO):int(DATASET_SIZE*(TRAIN_RATIO+VALIDATION_RATIO))]
y_val=y[int(DATASET_SIZE*TRAIN_RATIO):int(DATASET_SIZE*(TRAIN_RATIO+VALIDATION_RATIO))]
print(X_val.shape)
print(y_val.shape)
```

```python
val_dataset = tf.data.Dataset.from_tensor_slices((X_val,y_val))
#if data if not already shuffled we can shuffle using this data API
val_dataset = train_dataset.shuffle(buffer_size=8,reshuffle_each_iteration=True).batch(32).prefetch(tf.data.AUTOTUNE)
```

```python
X_test=X[int(DATASET_SIZE*(TRAIN_RATIO+VALIDATION_RATIO)):]
y_test=y[int(DATASET_SIZE*(TRAIN_RATIO+VALIDATION_RATIO)):]
print(X_test.shape)
print(y_test.shape)
```

```python
test_dataset = tf.data.Dataset.from_tensor_slices((X_test,y_test))
#if data if not already shuffled we can shuffle using this data API
test_dataset = train_dataset.shuffle(buffer_size=8,reshuffle_each_iteration=True).batch(32).prefetch(tf.data.AUTOTUNE)
```

## Getting familiar with Normalization

```python
normalizer = Normalization(axis=-1,mean=5,variance=4)
x_normalized = tf.constant([[3,4,5,6,7],
                            [4,5,6,7,8]])
normalizer(x_normalized)
```

### normalization = (x-mean)/std_dev
### std_dev = sqrt(variance)

```python
 adapt_normalizer = Normalization()
 adapt_normalizer.adapt(x_normalized)
 adapt_normalizer(x_normalized)
 #adaptive Normalizer is used for each column as axis=-1 by default
```

### Comming back to the project

```python
normalizer = Normalization()
normalizer.adapt(X_train)
normalizer(X_train)[:5]
```

```python
X_train[:5]
```

## Starting with the ML Model

```python
model = tf.keras.Sequential([
    InputLayer(input_shape=(8,)),
    normalizer,
    Dense(128,activation='relu'),
    Dense(128,activation='relu'),
    Dense(128,activation='relu'),
    Dense(1),
])
model.summary()
```

Sequential API,
Functional API,
Subclassing method

```python
#we can build the same model by writing it like this
# model = tf.keras.Sequential()
# model.add(normalizer)
# model.add(Dense(1))
```

```python
tf.keras.utils.plot_model(model,to_file="model.png",show_shapes=True)
```

## Error Sanctioning

### Different Loss Functions(MeanSquaredError,MeanAbsoluteError,Huber)

```python
# model.compile(loss=MeanSquaredError())
model.compile(optimizer = Adam(learning_rate=0.1) , loss=MeanAbsoluteError(),metrics=RootMeanSquaredError())
# model.compile(loss=Huber(delta = 0.2))
```

## Training And Optimization

```python
history = model.fit(train_dataset,validation_data=val_dataset,epochs=100,verbose=1)#epochs here means no. of times we will update our weights and verbose is the number of outputs from the tranining step
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','val_loss'])
plt.show()
```

## Performance Measurement

```python
plt.plot(history.history['root_mean_squared_error'])
plt.plot(history.history['val_root_mean_squared_error'])
plt.title('Model Performance')
plt.ylabel('rmse')
plt.xlabel('epoch')
plt.legend(['train','val'])
plt.show()
```

```python
model.evaluate(X_test,y_test)
```
